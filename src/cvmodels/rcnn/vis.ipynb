```python
import matplotlib.pyplot as plt

# Define the epoch numbers
epochs = range(1, 11)

# Define the train and validation losses
train_losses = [0.0661, 0.0627, 0.0659, 0.0651, 0.0651, 0.0651, 0.0623, 0.065, 0.0649, 0.0651]
validation_losses = [0.0627, 0.0629, 0.066, 0.0627, 0.0628, 0.0626, 0.0629, 0.0625, 0.0646, 0.0653]

# Plotting the train and validation losses
plt.figure(figsize=(10, 5))
plt.plot(epochs, train_losses, 'b', label='Train loss')
plt.plot(epochs, validation_losses, 'r', label='Validation loss')
plt.title('Train and Validation Losses')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.xticks(epochs)
plt.legend()
plt.grid(True)
plt.show()
```
